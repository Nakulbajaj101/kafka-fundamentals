```python
from confluent_kafka import Producer
import json
```

```python
producer_config = {
    "bootstrap.servers": "localhost:9092"
}
```

```python
producer = Producer(producer_config)
```

```python
orders = {"order_id": "A0225", "user_id": 7}
```

```python
def ack(err, msg):
    if err:
        print(err)
```

```python
producer.produce("orders", 
                 key=orders["order_id"], 
                 value=json.dumps(orders).encode('utf-8'),
                 callback=ack)
producer.poll(0)
```
